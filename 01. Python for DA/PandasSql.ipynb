```python
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
q1 = "SELECT usps, count(1) FROM df_pop GROUP BY usps"
```

```python
q2 = "SELECT * FROM df_pop WHERE usps='CA'"
```

```python
pysqldf(q1)
```

```python
pysqldf(q2)
```

```python
q3 = "select count(1) as count, sum(pop2022) as sum, avg(pop2022) as avg, max(pop2022) as max, min(pop2022) as min  from df_pop"
```

```python
pysqldf(q3)
```

```python
q4 = "select class, count(1) as count, sum(pop2022) as sum from (select (case when pop2022 > 1000000 then 'high' else 'not high' end) as class, pop2022 from df_pop) group by class"
```

```python
pysqldf(q4)
```

```python
kelas = []
for index, row in df_pop.iterrows():
    if row['pop2022'] > 1000000:
        kelas.append('high')
    else:
        kelas.append('not high')
```

```python
df_pop['class'] = kelas
```

```python
df_pop.head()
```

```python
df_pop.groupby(['class']).count()
```
