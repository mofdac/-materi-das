# Latihan 2 
Sumber data : Sales_June_2019.csv dan Sales_July_2019.csv 

```python
sales_june = pd.read_csv("https://github.com/mofdac/-materi-das/raw/main/01.%20Python%20for%20DA/latihan/Sales_June_2019.csv")
sales_july = pd.read_csv("https://github.com/mofdac/-materi-das/raw/main/01.%20Python%20for%20DA/latihan/Sales_July_2019.csv")
print(sales_june.shape)
display(sales_june.head())
print(sales_july.shape)
display(sales_july.head())
```

## 6. tampilkan berapa data null untuk masing-masing kolom

```python
print(f"Number of rows in June sales data with null value: {sales_june.loc[sales_june['Order ID'].isnull()].shape[0]} rows")

print(f"Number of rows in July sales data with null value: {sales_july.loc[sales_july['Order ID'].isnull()].shape[0]} rows")
```

## 7. Gabungkan kedua data diatas dan beri nama "df_sales" dataframe barunya

```python
df_sales = pd.concat([sales_june,sales_july]).reset_index(drop=True)
df_sales
```

## 8. cek dan hapus data dengan nilai null

```python
df_sales.loc[df_sales['Order ID'].isnull()]
```

```python
# Drop null row and reset index
df_sales = df_sales.dropna().reset_index(drop=True)
df_sales.info()
```

## 9. Tampilkan 10 produk dengan Nilai Penjualan Tertinggi Juni s.d. Juli (jangan lupa bikin kolom sales dan reset_index)

```python
#Calculate sale value per transaction
df_sales['Sales'] = df_sales['Quantity Ordered']*df_sales['Price Each']
df_sales
```

```python
df_sales_top_10_product = df_sales[['Product','Sales']].groupby(['Product']).sum().reset_index().\
                            sort_values('Sales', ascending=False).reset_index(drop=True).iloc[:10]
df_sales_top_10_product
```

## 10. Export soal 9 ke dalam format csv, csv.nya diberi nama "top_10_product_sales.csv"

```python
df_sales_top_10_product.to_csv('top_10_product_sales.csv',sep=';',index=False)
```

# Latihan 3

```python
# Sumber data
url_diabetes_prevalence_data = 'https://raw.githubusercontent.com/mihoku/MoF-Training-Python-for-Data-Analytics/main/Day%202/DM_Province_Data.csv'

url_konsumsi_minuman_kemasan_data = 'https://raw.githubusercontent.com/mihoku/MoF-Training-Python-for-Data-Analytics/main/Day%202/Data_Konsumsi_Minuman_Kemasan.csv'

url_kepadatan_penduduk_data = 'https://raw.githubusercontent.com/mihoku/MoF-Training-Python-for-Data-Analytics/main/Day%202/Data_Kepadatan_Penduduk_Provinsi.csv'
```

## 11. Data konsumsi minuman kemasan adalah data di tingkat kabupaten/kota. Buat dataframe baru yang merupakan data konsumsi minuman kemasan per provinsi.

```python
# baca file csv untuk konsumsi minuman kemasan
konsumsi_minuman_kemasan_data = pd.read_csv(url_konsumsi_minuman_kemasan_data, sep=';', dtype={'Kode_Provinsi':str,'Kode_Kabupaten_Kota':str})
konsumsi_minuman_kemasan_data.head()
```

```python
# Urutan syntax: pilih kolom-->groupby-->sum-->reset index
konsumsi_minuman_kemasan_data_province = konsumsi_minuman_kemasan_data[['Kode_Provinsi','Konsumsi_Minuman_Kemasan']].groupby(['Kode_Provinsi']).sum().reset_index()

konsumsi_minuman_kemasan_data_province.head()
```

## 12. Gabungkan data prevalensi diabetes, data konsumsi minuman kemasan, dan kepadatan penduduk (semuanya dalam provinsi), menjadi satu dataset

```python
# baca file data prevalensi diabetes
diabetes_prevalence_data = pd.read_csv(url_diabetes_prevalence_data, sep='#', dtype={'KDPROV':str})
diabetes_prevalence_data
```

```python
#baca file data kepadatan penduduk
kepadatan_penduduk_data = pd.read_csv(url_kepadatan_penduduk_data, sep=';')
kepadatan_penduduk_data.head()
```

```python
# merge dataset prevalensi diabetes dan dataset kepadatan penduduk
diabetes_prevalence_kepadatan_merged = pd.merge(left=diabetes_prevalence_data, right=kepadatan_penduduk_data, \
                                                how='left', left_on=['NM_PROV'], \
                                                right_on=['Nama_Provinsi'])[['KDPROV','NM_PROV','Prevalensi_DM','Kepadatan_Penduduk']]
diabetes_prevalence_kepadatan_merged.head()
```

```python
# merge ketiga dataset
dataset_merged = pd.merge(left=diabetes_prevalence_kepadatan_merged, right=konsumsi_minuman_kemasan_data_province, \
                                                how='left', left_on=['KDPROV'], \
                                                right_on=['Kode_Provinsi'])\
                                                [['KDPROV','NM_PROV','Prevalensi_DM','Kepadatan_Penduduk','Konsumsi_Minuman_Kemasan']]
dataset_merged
```
